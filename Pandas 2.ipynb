### na_values - to handle missing values

```python
df= pd.read_csv('nyc_weather.csv', na_values=['not available','na', '-1'])
df.head(15)
```

```python
df
```

```python
df.fillna()
```

```python
# if we want to fill the some specific name to some then we will use this method 

df.fillna({'WindSpeedMPH':0, 'Events':'No Event'}).head(20)
```

```python
df['WindSpeedMPH'].fillna(method="ffill", limit=5)
```

```python

df['WindSpeedMPH'].fillna(df['WindSpeedMPH'].mean())
```

```python
# this method is use to drop all the rows and columns where some value and not defined like NaN, 0, etc


df.dropna()
```

```python
# deleting columns which contain missing entries 
# in this windspeed and events column will not be appear


df.dropna(axis = 1)
```
