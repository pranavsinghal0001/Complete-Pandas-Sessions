### Display the most expensive car company name and price

```python
# in this we firstly describe the company and price from the cars dataframe 
# and then max the price name from the company

cars[['company','price']][cars.price==cars.price.max()]
```

### Find each company highest price car

```python
cname = cars.groupby('company')
cname[['company','price']].max()
```

### find the average mileage of each car making company

```python
cname = cars.groupby('company')
cname[['company','average-mileage']].mean()
```
