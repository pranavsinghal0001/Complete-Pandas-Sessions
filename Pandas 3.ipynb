### Using loc to fetch few rows and column from dataframe

```python
df.head(5)
```

```python
# if we want fetch the above three rows then we will use this 

df.loc[0:2,:]
```

```python
# if we want only above three rows and column from est to humidity

df.loc[0:2,'EST':'Humidity']
```

```python
# if we want some specific rows from the dataframe and column from est to humidity

df.loc[[5,12,15],'EST':'Humidity']
```

```python
# if we want the temperature above than 45 from the column est to humidity  

df.loc[df.Temperature>45,'EST':'Humidity']
```

```python
# if we want the temperature greater than 45 and the column are est temperature and dewpoint

df[['EST', 'Temperature','Humidity']][df.Temperature>45]
```

```python
# if we want maximum temperature we will use this

df.loc[df.Temperature==df.Temperature.max(),'EST':'DewPoint']
```

### Grouping 

```python
g = df.groupby('Events')
```

```python
g.get_group('Rain')
```

```python
for k, v in g:
    print('Group Name :,', k)
    print('--------------------')
    print(v)
```
